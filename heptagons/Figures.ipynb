```python
r = (0,-1,1)
s = (-1,1,0)
t = (1,-1,0)
u = (-1,1,0)

def scale( s, v ) :
    x, y = v
    return ( times( s, x ), times( s, y ) )

def rotate(v) :
    x, y = v
    return ( plus( times(r,x), times(t,y) ), plus( times(s,x), times(u,y) ) )

def plusv( v1, v2 ) :
    h1, h2 = v1
    h3, h4 = v2
    return ( plus( h1, h3 ), plus( h2, h4 ) )

v0 = ( rho_sigma, zero )
v1 = ( sigma_inv, sigma_inv )

origin = ( zero, zero )
```

```python
v = v0
for i in range( 15 ):
    v = rotate( v )
    print( v )
```

```python
def render( v ):
    x, y = v
    return [ asReal(x), asReal(y) ]
```

```python

def heptagonVerts(v,skip=1):
    result = []
    vi = v
    for i in range(7):
        result .append( render( vi ) )
        for j in range(skip):
            vi = rotate( vi )
    return result

v0 = ( rho_sigma, zero )
heptagon1 = heptagonVerts( v0 )
heptagon2 = heptagonVerts( plusv( v0, v1 ) )
heptagon3 = heptagonVerts( plusv( plusv( v0, v1 ), v1 ) )
heptagon4 = heptagonVerts( ( neg( rho_sigma ), zero ) )

heptagram1 = heptagonVerts( v0, 2 )
heptagram2 = heptagonVerts( plusv( v0, v1 ), 3 )
```

A tiny bit of trigonometry and linear algebra yields a matrix to perform the necessary shear transformation:

$$\begin{bmatrix}1 & \frac{1}{2\sigma} \\0 & \sin\frac{3}{7}\pi\end{bmatrix}$$

```python
def render(v):
    x = asReal( v[0] )
    y = asReal( v[1] )
    A = math.sin( (3.0/7.0) * math.pi )
    return [ x + y/(2*sigma_value), A*y ]
```

```python
def ngonPath( edge, n, skip=1, start=origin ):
    result = []
    vi = plusv( start, edge )
    for i in range(n):
        result .append( render( vi ) )
        for j in range(skip):
            edge = rotate( edge )
        vi = plusv( vi, edge )
    return result
```

```python
one_edge = ( sigma, zero )
heptagon_path = ngonPath( one_edge, 7 )

rho_edge = plusv( one_edge, rotate( one_edge ) )
heptagram_rho_path = ngonPath( rho_edge, 7, 2 )

sigma_edge = ( zero, plus( plus( sigma, rho ), one ) )
heptagram_sigma_path = ngonPath( sigma_edge, 7, 4 )

inner_origin = ( plus( sigma, neg( rho) ), rho )
small_edge = ( rho_inv, zero )
small_rho_edge = plusv( small_edge, rotate( small_edge ) )
small_sigma_edge = ( zero, (-1,0,1) )

heptagon_path_shrunk = ngonPath( small_edge, 7, 1, inner_origin )

heptagram_rho_path_shrunk = ngonPath( small_rho_edge, 7, 2, inner_origin )

heptagram_sigma_path_shrunk = ngonPath( small_sigma_edge, 7, 4, inner_origin )
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.add_patch( drawPolygon( heptagon_path,'#dd0000' ) )
ax.add_patch( drawPolygon( heptagram_rho_path,'#990099' ) )
ax.add_patch( drawPolygon( heptagram_sigma_path,'#0000dd' ) )
ax.add_patch( drawPolygon( heptagon_path_shrunk,'#dd0000' ) )
ax.add_patch( drawPolygon( heptagram_rho_path_shrunk,'#990099' ) )
ax.add_patch( drawPolygon( heptagram_sigma_path_shrunk,'#0000dd' ) )
ax.set_xlim(-1.5,3.7)
ax.set_ylim(-0.2,5)
fig.text(0.52, 0.91, 'A',
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes, fontsize=14, weight='bold' )

if saveFigs==1:
    ax.axis('off')
    fig.savefig("nestedHeptagrams.png")
```

```python

def render(v):
    x = asReal( v[0] )
    y = asReal( v[1] )
    A = math.sin( (3.0/7.0) * math.pi )
    x = x + y/(2*sigma_value)
    y = A*y
    return [ -y, x ]

def renderCentroid( v1, v2 ):
    p1 = render( v1 )
    p2 = render( v2 )
    return [ (p1[0]+p2[0])/2, (p1[1]+p2[1])/2 ]

def cyclicPath( edge, skip=1, start=origin ):
    result = []
    vi = plusv( start, edge )
    for i in range(7):
        result .append( vi )
        for j in range(skip):
            edge = rotate( edge )
        vi = plusv( vi, edge )
    return result

heptagon_path = cyclicPath( one_edge )
heptagram_rho_path = cyclicPath( rho_edge, 2 )
heptagram_sigma_path = cyclicPath( sigma_edge, 4 )


def drawPolygon( polygonVerts, color ):
    n = len( polygonVerts )
    codes = [ Path.MOVETO ]
    verts = []
    verts .append( render( polygonVerts[ 0 ] ) )
    for i in range(1,n+1):
        codes.append ( Path.LINETO )
        verts.append ( render( polygonVerts[ i % n ] ) )
    path = mpath.Path( verts, codes )
    return mpatches.PathPatch( path, facecolor='none', edgecolor=color )

def xAxis( v, color ):
    start = render( v )
    codes = [ Path.MOVETO ]
    verts = []
    verts .append( [ start[0]-2, start[1]-0.01 ] )
    codes .append ( Path.LINETO )
    verts .append ( [ start[0]+6, start[1]-0.01 ] )
    path = mpath.Path( verts, codes )
    return mpatches.PathPatch( path, facecolor='none', edgecolor=color )

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.add_patch( drawPolygon( heptagon_path,'#dd0000' ) )
ax.add_patch( drawPolygon( heptagram_rho_path,'#990099' ) )
ax.add_patch( drawPolygon( heptagram_sigma_path,'#0000dd' ) )
ax.add_patch( xAxis( heptagon_path[ 3 ],'#000000' ) )

def point( p ) :
    ax.add_patch( Circle( (p[0], p[1]), 0.04 ) )
point( render( heptagon_path[ 0 ] ) )
point( render( heptagon_path[ 2 ] ) )
point( render( heptagon_path[ 3 ] ) )
point( renderCentroid( heptagon_path[ 2 ],  heptagon_path[ 4 ] ) )
point( renderCentroid( heptagon_path[ 0 ],  heptagon_path[ 6 ] ) )

def label( s, x, y ):
    fig.text( x, y, s,
         horizontalalignment='center',
         verticalalignment='center',
         transform = ax.transAxes, fontsize=14, weight='bold' )

label( 'A', 0.18, 0.55 )
label( 'B', 0.86, 0.69 )
label( 'C', 0.86, 0.55 )
label( 'D', 0.34, 0.82 )
label( 'E', 0.35, 0.55 )

ax.set_xlim(-5.5,0.5)
ax.set_ylim(-2,4)
if saveFigs==1:
    ax.axis('off')
    fig.savefig("findingConstants.png")
```

```python
def drawPolygonSubset(polygonVerts,todraw,color):
    n = len( polygonVerts )
    codes = [ Path.MOVETO ]
    verts = []
    verts .append( polygonVerts[ 0 ] )
    for i in range(1,n+1):
        if i in todraw:
            codes.append ( Path.LINETO )
        else:
            codes.append ( Path.MOVETO )
        verts.append ( polygonVerts[ i % n ] )
    path = mpath.Path( verts, codes )
    return mpatches.PathPatch( path, facecolor='none', edgecolor=color )

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.add_patch( drawPolygon( heptagon_path,'#dd0000' ) )
ax.add_patch( drawPolygonSubset( heptagram_rho_path, set([4,5]), '#990099' ) )
ax.add_patch( drawPolygonSubset( heptagram_sigma_path, set([7,1,4]), '#0000dd' ) )
ax.set_xlim(-1.5,3.7)
ax.set_ylim(-0.2,5)
ax.axis('off')
fig.savefig("heptagonSimilarTriangles.png")
```

```python

def renderOffset(v):
    x = asReal( plus( v[0], (0,2,1) ) )
    y = asReal( plus( v[1], (0,1,0) ) )
    A = math.sin( (3.0/7.0) * math.pi )
    return [ x + y/(2*sigma_value), A*y ]

def nautilus( n ):
    result = []
    def shrink(v):
        return scale( (-1,1,0), v )
    p1 = ( sigma, (-1,0,1) )
    p2 = ( zero, sigma )
    for i in range(n):
        result .append( renderOffset( origin ) )
        result .append( renderOffset( p1 ) )
        result .append( renderOffset( p2 ) )
        p1 = p2
        p2 = rotate( shrink( p2 ) )
    return result

fig = plt.figure(figsize=(11,6))
ax = fig.add_subplot(111)
ax.add_patch( drawPolygon( heptagon_path,'#dd0000' ) )
ax.add_patch( drawPolygon( heptagram_rho_path,'#990099' ) )
ax.add_patch( drawPolygon( heptagram_sigma_path,'#0000dd' ) )
ax.add_patch( drawPolygon( heptagon_path_shrunk,'#dd0000' ) )
ax.add_patch( drawPolygon( heptagram_rho_path_shrunk,'#990099' ) )
ax.add_patch( drawPolygon( heptagram_sigma_path_shrunk,'#0000dd' ) )
ax.add_patch( drawPolygon( nautilus( 22 ), '#0044aa' ) )
ax.set_xlim(-2,9)
ax.set_ylim(-0.5,5.5)
ax.axis('off')
fig.savefig("heptagonSampler.png")
```
